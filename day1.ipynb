```python
file = open('day1_input.txt', 'r')
text_lines=file.readlines()
len(text_lines)
```

```python
left=[]
right=[]
for line in text_lines:
    nums=line.split()
    left.append(int(nums[0]))
    right.append(int(nums[1]))
```

```python
left=sorted(left)
right=sorted(right)
s=0
for i in range(len(left)):
    s+=abs(left[i]-right[i])
```
